```julia
using DifferentialEquations
using Plots

#     
k = 8.1

phi = 3*pi/4

#  .
ode_fn(r, p, t) = r/sqrt(9.24)

f(t) = t*tan(phi)

t_begin = 0.0
t_end = pi
t_span1 = (t_begin, t_end)

t_span2 = (-pi, 0.0)

boot_values = fill(sqrt(2)/2, 4)

# 
r_01 = k/2.2
r_02 = k/3.2

prob1 = ODEProblem(ode_fn, r_01, t_span1)
sol1 = solve(prob1, Tsit5(), reltol=1e-8, abstol=1e-8)

prob2 = ODEProblem(ode_fn, r_02, t_span2)
sol2 = solve(prob2, Tsit5(), reltol=1e-8, abstol=1e-8)

plot(proj=:polar,
     sol1.t,
     linewidth = 2,
     title = "  #1",
     label = " ",
     color =:red,
     legend = true)

plot!(boot_values, collect(0:3), linewidth = 2, label = "  ", 
      color =:blue,
      legend=true)
```

```julia
plot(proj=:polar,
     sol2.t,
     linewidth = 2,
     title = "  #2",
     label = " ",
     color =:red,
     legend = true)

plot!(boot_values, collect(0:3), linewidth = 2, label = "  ", 
      color =:blue,
      legend=true)
```
