```julia
using Plots
using DifferentialEquations
```

```julia
M1 = 4.4
M2 = 4
p_cr = 10.5
N = 28
q = 1
1 = 16
2 = 25
p1 = 7.2
p2 = 5.1

a1 = p_cr/((1^2) * (p1^2) * N *q)
a2 = p_cr/((2^2) * (p2^2) * N *q)
b = p_cr/((1^2) * (p1^2) * (2^2) * (p2^2) * N *q)
c1 = (p_cr-p1)/(1 * p1)
c2 = (p_cr - p2)/(2 * p2)

function ode_fn_1(du, u, p, t)
    x,y = u 
    du[1] = x - (b/c1)*x*y - (a1/c1)*(x^2)
    du[2] = (c2/c1)*y - (b/c1)*x*y - (a2/c1)*(y^2)
end

init_val = [M1, M2]

t_begin = 0
t_end = 30
tspan = (t_begin, t_end)

prob = ODEProblem(ode_fn_1, init_val, tspan)
sol = solve(prob, Tsit5(), reltol=1e-16, abstol=1e-16)

x_sol = [u[1] for u in sol.u]
y_sol = [u[2] for u in sol.u]

plot(sol.t * c1, x_sol, 
    linewidth = 2,
    title = "   ",
    label = "M1",
    legend = true)

plot!(sol.t * c1, y_sol, 
    linewidth = 2,
    title = "   ",
    label = "M2",
    legend = true)
```
